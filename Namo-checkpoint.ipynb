```python
df1=df.groupby(['Election Year'])
```

```python
df1.first().count()
```

# For The Feb-2005

```python
feb=df1.get_group('2005-Feb')
```

```python
feb
```

```python
feb['Party'].unique()
```

```python
febp=feb.groupby(['Party']).sum()
```

```python
febp
```

```python
result1=febp.sort_values(by=['Votes'], ascending=False)
```
